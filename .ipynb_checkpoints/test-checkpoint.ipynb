```python
def DI(temp, humid):
    '''
    disscomfort index : 불쾌지수
    '''
    return 0.81 * temp +0.01 * humid * (0.99*temp - 14.3)+ 46.3
```

```python
train['DI'] = DI(train['temperature'], train['humidity'])
train['feel_DI'] = train['DI'].map(lambda x : 1 if x > 80 else 0)
```

```python
X_train,X_test,y_train,y_test = train_test_split(train,y,test_size=0.2,random_state=1234)
```

```python
cat = CatBoostRegressor()
```

```python
cat.fit(X_train, y_train)
```

```python
# metric
def SMAPE(true, pred):
    return np.mean((np.abs(true-pred))/(np.abs(true) + np.abs(pred)))*100
```

```python
pred = cat.predict(X_test)

SMAPE(y_test, pred)
```

```python
def plot_feature_importance(importance, names, model_type):
    
    feature_importance = np.array(importance)
    feature_names = np.array(names)
    
    data={'feature_names':feature_names,'feature_importance':feature_importance}
    fi_df = pd.DataFrame(data)
    
    fi_df.sort_values(by=['feature_importance'], ascending=False,inplace=True)

    plt.figure(figsize=(10,8))

    sns.barplot(x=fi_df['feature_importance'], y=fi_df['feature_names'])

    plt.title(model_type + ' Feature Importance')
    plt.xlabel('Feature Importance')
    plt.ylabel('Feature Names')
```

```python
plot_feature_importance(cat.get_feature_importance(),train.columns,"CATBOOST")
```
